```python
import numpy as np
import pandas as pd
import xarray as xr

def import_data(x_file, y_file):
    # read the data from the files into separate dataframes
    x_df = pd.read_csv(x_file, header=None)
    y_df = pd.read_csv(y_file, header=None)

    # merge the two dataframes into one based on the index
    data_df = pd.concat([x_df, y_df], axis=1)

    # add column names to the dataframe
    data_df.columns = ['x', 'y']

    return data_df

data_df = import_data('TimeSteps_h.txt', 'At18_Diff.txt')

print(data_df)
```
